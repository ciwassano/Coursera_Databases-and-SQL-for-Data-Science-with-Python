```python
schools_df = pd.read_sql('SELECT * FROM Coursera.dbo.ChicagoPublicSchools', conexao)
crimes_df = pd.read_sql('SELECT * FROM Coursera.dbo.ChicagoCrimeData', conexao)
census_df = pd.read_sql('SELECT * FROM Coursera.dbo.ChicagoCensusData', conexao)
display(schools_df)
display(crimes_df)
display(census_df)
```

### Problem 1: Find the total number of crimes recorded in the CRIME table.

```python
# Por pandas
qtd_crimes = len(crimes_df)
print('Número de crimes = {}'.format(qtd_crimes))
```

```python
#por SQL
qtd_crimes_sql = pd.read_sql('SELECT COUNT(*) FROM ChicagoCrimeData', conexao)
print('Número de crimes = {}'.format(qtd_crimes_sql.iloc[0][0]))
```

### Problem 2: List community areas with per capita income less than 11000.

```python
#por pandas

com_poor_areas = census_df['COMMUNITY_AREA_NAME'][census_df['PER_CAPITA_INCOME'] < 11000]
print('Community areas with per capita income less than 11000:')
for i in com_poor_areas:
    print(i)
```

```python
#por SQL
com_poor_areas_sql = pd.read_sql('SELECT COMMUNITY_AREA_NAME FROM ChicagoCensusData WHERE PER_CAPITA_INCOME < 11000', conexao )
display(com_poor_areas_sql)
```

### Problem 3: List all case numbers for crimes involving minors?

```python
# por pandas
crimes_minor=crimes_df['CASE_NUMBER'][crimes_df['DESCRIPTION'].str.contains('MINOR')]
print(crimes_minor)
```

```python
crimes_minor_sql = pd.read_sql("SELECT CASE_NUMBER FROM ChicagoCrimeData WHERE DESCRIPTION like '%minor%'", conexao)
display(crimes_minor_sql)
```

### Problem 4: List all kidnapping crimes involving a child?(children are not considered minors for the purposes of crime analysis)

```python
# por pandas
crimes_kidnapchild = crimes_df['CASE_NUMBER'][(crimes_df['PRIMARY_TYPE'].str.contains('KIDNAPPING')) & crimes_df['DESCRIPTION'].str.contains('CHILD')]
display(crimes_kidnapchild)
```

```python
#por SQL
crimes_kidnapchild_sql = pd.read_sql("SELECT CASE_NUMBER FROM ChicagoCrimeData WHERE PRIMARY_TYPE like '%kidnapping%' AND DESCRIPTION like '%child%'", conexao)
display(crimes_kidnapchild_sql)
```

### Problem 5: What kind of crimes were recorded at schools?

```python
#por pandas

crimes_school = crimes_df[['PRIMARY_TYPE', 'DESCRIPTION']][crimes_df['LOCATION_DESCRIPTION'].str.contains('SCHOOL')]
display(crimes_school['PRIMARY_TYPE'].unique())
```

```python
#por SQL

crimes_school_sql = pd.read_sql("SELECT DISTINCT(PRIMARY_TYPE) FROM ChicagoCrimeData \
                                WHERE LOCATION_DESCRIPTION like '%school%'", conexao)
display(crimes_school_sql)
```

### Problem 6: List the average safety score for all types of schools.

```python
sft_score = pd.read_sql("SELECT Elementary_Middle_or_High_School, AVG(SAFETY_SCORE) \
                        AS AVERAGE_STY_SCORE FROM ChicagoPublicSchools GROUP BY Elementary_Middle_or_High_School", conexao)
display(sft_score)
```

```python
#por python
schools_df['sfty_score'] = pd.to_numeric(schools_df["SAFETY_SCORE"])
school_type_df = schools_df.groupby("Elementary_Middle_or_High_School")
mean_school = school_type_df["sfty_score"].mean()
display(mean_school)
```

```python
#por SQL
cursor.execute("SELECT Elementary_Middle_or_High_School, AVG(SAFETY_SCORE) \
               AS AVERAGE_STY_SCORE FROM ChicagoPublicSchools GROUP BY Elementary_Middle_or_High_School")
cursor.fetchall()
```

### Problem 7: List 5 community areas with highest % of households below poverty line.

```python
#por sql
high_five_comm_hsehld_sql = pd.read_sql('SELECT top 5 COMMUNITY_AREA_NAME FROM ChicagoCensusData \
                                        ORDER BY PERCENT_HOUSEHOLDS_BELOW_POVERTY DESC', conexao)
display(high_five_comm_hsehld_sql)
```

### Problem 8: Which community area(number) is most crime prone?

### Problem 9: Use a sub-query to find the name of the community area with highest hardship index.

```python
#por sql
comm_crime = pd.read_sql("SELECT COMMUNITY_AREA_NUMBER, COMMUNITY_AREA_NAME, HARDSHIP_INDEX  FROM ChicagoCensusData \
                         WHERE HARDSHIP_INDEX =(SELECT MAX(HARDSHIP_INDEX) FROM ChicagoCensusData)", conexao)
display(comm_crime)

#ou
cursor.execute("SELECT COMMUNITY_AREA_NUMBER, COMMUNITY_AREA_NAME, HARDSHIP_INDEX  FROM ChicagoCensusData \
               WHERE HARDSHIP_INDEX =(SELECT MAX(HARDSHIP_INDEX) FROM ChicagoCensusData)")
cursor.fetchall()
```

### Problem 10: Use a sub-query to determine the Community Area Name with most number of crimes?

```python
area_most_crimes = pd.read_sql("select COMMUNITY_AREA_NUMBER, COMMUNITY_AREA_NAME FROM ChicagoCensusData \
                               WHERE COMMUNITY_AREA_NUMBER IN (SELECT TOP 2 COMMUNITY_AREA_NUMBER FROM ChicagoCrimeData \
                               GROUP BY COMMUNITY_AREA_NUMBER ORDER BY COUNT(*) DESC)", conexao)
display(area_most_crimes)
```

```python
#most_crimes = pd.read_sql("SELECT COMMUNITY_AREA_NAME FROM ChicagoCensusData WHERE ")

df_mostcrimes = crimes_df.groupby(['COMMUNITY_AREA_NUMBER']).count()
#display(df_mostcrimes)
total = df_mostcrimes['CASE_NUMBER'].sum()
#print(total)
#print(df_mostcrimes['CASE_NUMBER'].max())

df_crimesbycommunity = df_mostcrimes.merge(census_df, on = "COMMUNITY_AREA_NUMBER")
df_crimesbycommunity = df_crimesbycommunity[['COMMUNITY_AREA_NUMBER', 'CASE_NUMBER', 'COMMUNITY_AREA_NAME']]
#display(df_crimesbycommunity)

area_most_crimes = df_crimesbycommunity.nlargest(1, 'CASE_NUMBER')
display(area_most_crimes)
```
